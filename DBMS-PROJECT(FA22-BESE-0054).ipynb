```python
# DBMS PROJECT (FA22-BESE-0054)(ASAD RAZA)

import pyodbc

Connector = pyodbc.connect('Driver={SQL Server};'
                          'Server=LAPTOP-90L04CRC\\SYEDASADRAZA;'
                          'Database=AIRPORTMANGEMENT;'
                          'Trusted_Connection=yes;')

def list_airlines():
    cursor = Connector.cursor()
    cursor.execute('SELECT * FROM Airline')
    for row in cursor:
        print(row)
    cursor.close()

# Function to add a new airline
def add_airline():
    # Prompt user for airline details
    airline_name = input("Enter airline name: ")

    # Insert new airline into Airline table
    cursor = Connector.cursor()
    cursor.execute('INSERT INTO Airline (AirlineName) VALUES (?)', (airline_name,))
    Connector.commit()
    cursor.close()
    print("Airline added successfully.")

# Function to list all airplanes
def list_airplanes():
    cursor = Connector.cursor()
    cursor.execute('SELECT * FROM Airplane')
    for row in cursor:
        print(row)
    cursor.close()

# Function to add a new airplane
def add_airplane():
    # Prompt user for airplane details
    model_number = input("Enter model number: ")
    registration_number = input("Enter registration number: ")
    passenger_capacity = input("Enter passenger capacity: ")
    airline_id = input("Enter airline ID: ")

    # Insert new airplane into Airplane table
    cursor = Connector.cursor()
    cursor.execute('''
        INSERT INTO Airplane (ModelNumber, RegistrationNumber, PassengerCapacity, AirlineID)
        VALUES (?, ?, ?, ?)
    ''', (model_number, registration_number, passenger_capacity, airline_id))
    Connector.commit()
    cursor.close()
    print("Airplane added successfully.")

# Function to list all flights
def list_flights():
    cursor = Connector.cursor()
    cursor.execute('SELECT * FROM Flight')
    for row in cursor:
        print(row)
    cursor.close()

# Function to add a new flight
def add_flight():
    # Prompt user for flight details
    flight_number = input("Enter flight number: ")
    departure_airport = input("Enter departure airport: ")
    destination_airport = input("Enter destination airport: ")
    departure_date_time = input("Enter departure date and time (YYYY-MM-DD HH:MM:SS): ")
    arrival_date_time = input("Enter arrival date and time (YYYY-MM-DD HH:MM:SS): ")
    airplane_id = input("Enter airplane ID: ")

    # Insert new flight into Flight table
    cursor = Connector.cursor()
    cursor.execute('''
        INSERT INTO Flight (FlightNumber, DepartureAirport, DestinationAirport, DepartureDateTime, ArrivalDateTime, AirplaneID)
        VALUES (?, ?, ?, ?, ?, ?)
    ''', (flight_number, departure_airport, destination_airport, departure_date_time, arrival_date_time, airplane_id))
    Connector.commit()
    cursor.close()
    print("Flight added successfully.")

# Function to list all passengers
def list_passengers():
    cursor = Connector.cursor()
    cursor.execute('SELECT * FROM Passenger')
    for row in cursor:
        print(row)
    cursor.close()

# Function to add a new passenger
def add_passenger():
    # Prompt user for passenger details
    first_name = input("Enter first name: ")
    last_name = input("Enter last name: ")
    email = input("Enter email: ")

    # Insert new passenger into Passenger table
    cursor = Connector.cursor()
    cursor.execute('''
        INSERT INTO Passenger (FirstName, LastName, Email)
        VALUES (?, ?, ?)
    ''', (first_name, last_name, email))
    Connector.commit()
    cursor.close()
    print("Passenger added successfully.")

# Function to list all bookings
def list_bookings():
    cursor = Connector.cursor()
    cursor.execute('SELECT * FROM Booking')
    for row in cursor:
        print(row)
    cursor.close()

# Function to add a new booking
def add_booking():
    # Prompt user for booking details
    passenger_id = input("Enter passenger ID: ")
    flight_id = input("Enter flight ID: ")

    # Insert new booking into Booking table
    cursor = Connector.cursor()
    cursor.execute('''
        INSERT INTO Booking (PassengerID, FlightID)
        VALUES (?, ?)
    ''', (passenger_id, flight_id))
    Connector.commit()
    cursor.close()
    print("Booking added successfully.")

# Main function to run the airport management system
def main():
    while True:
        print("\nAirport Management System")
        print("1. List all airlines")
        print("2. Add new airline")
        print("3. List all airplanes")
        print("4. Add new airplane")
        print("5. List all flights")
        print("6. Add new flight")
        print("7. List all passengers")
        print("8. Add new passenger")
        print("9. List all bookings")
        print("10. Add new booking")
        print("11. Exit")

        choice = input("Enter your choice (1-11): ")

        if choice == "1":
            list_airlines()
        elif choice == "2":
            add_airline()
        elif choice == "3":
            list_airplanes()
        elif choice == "4":
            add_airplane()
        elif choice == "5":
            list_flights()
        elif choice == "6":
            add_flight()
        elif choice == "7":
            list_passengers()
        elif choice == "8":
            add_passenger()
        elif choice == "9":
            list_bookings()
        elif choice == "10":
            add_booking()
        elif choice == "11":
            print("Exiting Airport Management System.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
